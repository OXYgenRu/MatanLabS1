```python
import math

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Дано:
Одномерное логистическое отображение:
$x_{n+1} = rx_n(1-x_n)$


### Easy level

Доказать:
$$
\forall n \in \mathbb{N}, \forall r \in (0,1]$ \\
0 < x_0 < 1 \Rightarrow 0 < x_n < 1
$$
Доказательство:
Докажем по принципу мат. индукции.
Проверим базу $0 < x_0 < 1$, верно по условию.
Пусть $0 < x_0 < 1 \Rightarrow 0 < x_n < 1$ верно для $n=k\quad k \in \mathbb{N}$, проверим для $ k+1$:
$$ 0 < x_k < 1 \quad\Rightarrow\quad\ -1 < -x_k < 0 \quad\Rightarrow\quad\ 0 < 1-x_k < 1 \quad\Rightarrow\quad\ 0 < x_k(1-x_k) < x_k $$
Из $r \in (0,1]$ Следует  $0 < rx_k(1-x_k) <= x_k(1-x_k) < x_k$ Тогда $ 0 < x_{k+1} < x_k < 1$
Предположение верно, тогда по принципу математической индукции доказано.

Рассмотрим как параметр r влияет на поведение функции $ x_{n+1} = rx_{n}(1-x_{n}) $
Если рассмотреть $ r=1\quad x_{n+1} = x_{n}(1-x_{n}) = x_{n} - x_{n}^2 $ - парабола с ветвями вниз,
тогда для $x_{n+1} = r(x_{n} - x_{n}^2) $, r - растягивает и сжимает параболу, для r > 1 растягивает, для r < 1 сжимает, для r = 1 ничего не меняет.

```python
def f(x_, r_):
    return r_ * x_ * (1 - x_)


list_r = np.linspace(0, 4, 5)
x = np.linspace(0, 1, 500)
plt.figure(figsize=(8, 6))

for r in list_r:
    x_next = f(x, r)
    plt.plot(x, x_next, label=f'r = {r}')

plt.xlabel('$x_n$')
plt.ylabel('$x_{n+1}$')
plt.title('Зависимость $x_{n+1}$ от $x_n$ для разных $r$')
plt.legend()
plt.grid(True)
plt.show()
```

Вариант лабораторной $N = 0$
$g(x_{n}) = rx_n(1-x_n)(2+x_n)\quad r \in [0,\frac{27}{2(7\sqrt{7}-10)}]$

```python
def g(x_, r_):
    return r_ * x_ * (1 - x_) * (2 + x_)


max_r = 27 / (2 * (7 * math.sqrt(7) - 10))
list_r = np.linspace(0, max_r, 5)

x = np.linspace(0, 1, 500)
plt.figure(figsize=(8, 6))

for r in list_r:
    x_next = g(x, r)
    plt.plot(x, x_next, label=f'r = {r}')

plt.xlabel('$x_n$')
plt.ylabel('$x_{n+1}$')
plt.title('Зависимость $x_{n+1}$ от $x_n$ для разных $r$')
plt.legend()
plt.grid(True)
plt.show()
```

Сходства с логическим отображением:
График похож на график параболы с ветвями вниз, что делает его схожим с графиком логического отображения.
При $x_n$ равном 0 и 1 $x_{n+1}$ равен 0 для обоих графиков.
Так же графики достигают максимума равного 1 при максимальном r.
Графики схожи так как максимальное r подобрано специально, чтобы $x_{n+1}$ оставался в промежутке от 0 до 1.

Различия:
График данного точечного отображения отличается от графика параболы, точка максимума смещена вправо, так как функция не задает параболу, а задает полином 3 степени.

### Normal level

Неподвижная точка, это точка для которой выполняется $x=f(x)$
Найдем все неподвижные точки логистического отображения
$x=rx(1-x)\quad\iff\quad x=rx-rx^2 \quad\iff\quad rx^2-rx+x=0\quad\iff\quad x(rx-r+1)=0$
Получаем одну из точек $x=0$
Найдем вторую $rx-r+1 = 0 \quad\iff\quad x= \frac{r-1}{r}\quad$ она существует при $r \neq 0$ и $x \in [0,1]$
$\frac{r-1}{r} \le 1 \quad\iff\quad r-1\le r \quad\iff\quad -1\le0\quad$ получаем, что это выполняется при любом $r$
$0 \le \frac{r-1}{r} \quad\iff\quad 0 \le r-1 \quad\iff\quad 1\le r\quad$ получаем вторая точка сщуествует при $1\le r$

Таким образом возможнные неподвижные точки можно найти по следующим формулам $x=0$ и $x= \frac{r-1}{r} = 1- \frac{1}{r}$

При $r<1$ существует единственная точка $x=0$, так как второй не существует для этого $r$.
При $r=1$ существует две точки $x=0$ и $x=1-\frac{1}{1}=0$, таким образом мы получаем единственную точку 0.
При $r>1$ существует две различные точки $x=0$ и $x=1-\frac{1}{r}$

Это отображение может иметь не более двух неподвижных точек, так как нахождение неподвижных точек сводится к решению 
уравнения второй степени, которое не может иметь более двух действительных корней.

Докажем, что при $x_0\in(0;1)$ и $r\in(0;1]$ последовательность $\{x_n\}$, заданная логистическим отображением, монотонно убывает.
Докажем, что $0< x_n < 1$ и $x_{n+1} < x_n$, что даст нам монотонность, а также ограниченность, которая понадобится в будущем.
Докажем с помощью математической индукции:
База для $n=0$, по условию $0<x_0<1$, тогда  $ 0 < x_0 < 1 \quad\Rightarrow\quad\-1 < -x_0 < 0 
\quad\Rightarrow\quad\ 0 < 1-x_0 < 1 \quad\Rightarrow\quad\ 0 < x_0(1-x_0) < x_0 < 1$ 
при $r\in(0;1]$  $\quad0 < rx_0(1-x_0) \le x_0(1-x_0)  < x_0 <1 \quad\Rightarrow\quad\ 0<x_1 < x_0 <1$
Переход индукции, пусть $0< x_n < 1$ и $x_{n+1} < x_n$ верно для $n=k \quad k\in \mathbb(N)$, получим утверждение для $n=k+1$.
Применив рассуждения использованные в доказательстве базы индукции из  $0< x_k < 1$ получим $0 < x_{k+1} < x_k < 1$.
Аналогично для $0 < x_{k+1} < 1$  
$0 < x_{k+1} < 1 \quad\Rightarrow\quad\-1 < -x_{k+1} < 0 \quad\Rightarrow\quad\ 0 < 1-x_{k+1} < 1 \quad\Rightarrow\quad\ 0 < x_{k+1}(1-x_{k+1}) < x_{k+1} < 1$ 
при $r\in(0;1]$  $\quad0 < rx_{k+1}(1-x_{k+1}) \le x_1(1-x_{k+1})  < x_{k+1} <1 \quad\Rightarrow\quad\ 0<x_{k+2} < x_{k+1} <1$
Получили, что утверждение выполнилось для $n = k + 1$, тогда переход доказан, тогда последовательность является ограниченной и монотонно убывающей.
Откуда по теореме Вейерштрасса об ограниченной и монотонной последовательности у $\{x_n\}$ есть передел.

```python

plt.figure(figsize=(8, 6))

r = 0.5
x_0 = 0.2
n = 50

x = [x_0]

for i in range(n):
    x.append(r * x[i] * (1 - x[i]))

plt.plot(range(n + 1), x, marker='o')
plt.xlabel('$n$')
plt.ylabel('$x_{n+1}$')
plt.title(f'График логистического отображения  при r = {r} и $x_0$={x_0}')
plt.grid(True)
plt.show()
```

Дано $r\in(2;3),\;x_{2n}>x^*,\;x_{2n+1} <x^*$, необходимо оценить монотонность последовательностей $\{x_{2n}\},\;\{x_{2n+1}\}$.
Пусть $f(x_n) = rx_n(1-x_n) = x_{n+1},\;y_n=x_{2n},\;z_n=x_{2n+1}$
Для оценки монотонности последовательностей необходимо оценить то, чему равны $y_{n+1}-y_n,\;z_{n+1}-z_n$
$y_{n+1}-y_n\:=\:x_{2n+2}-x_2n\:=\:f(f(x_2n))-x_2n,\quad z_{n+1}-z_n\:=\:x_{2n+3}-x_{2n+1}\:=\:f(f(x_{2n+1}))-x_{2n+1}$
Таким образом нужно научиться оценивать $f(f(x)) -x$
$f(f(x))-x\:=\:rf(x)(1-f(x))-x\:=\:r(rx(1-x))(1-rx(1-x))-x\:=\:-r^3x^4+2r^3x^3-(r^3+r^2)x^2+(r^2-1)x$ обозначим его за $P(x)$
Рассмотрим поведение $f(f(x))-x$ при $x=x^*$, так как $r\in(2;3)$ то воспользуемся уже известным фактом, существует две неподвижные точки,
а именно $x^*=0,\;x^*=1-\frac{1}{r}$, подставим $x^*$ в разность: $f(f(x^*))-x^*\;=\;f(x^*)-x^*\;=\;x^*-x^*\;=\;0$.
Причем данная по условию неподвижная точка $x^*$ равна именно $1-\frac{1}{r}$, потому что для $x^*=0$ не может выполняться $x_{2n+1} <x^*$.
Далее как $x^*$ будем понимать неподвижную точку $1-\frac{1}{r}$.
Тогда полученный многочлен $P(x)$ делится на $(x-0)$ и $(x-(1-\frac{1}{r})$.
Рассмотрим $(x-0)(x-(1-\frac{1}{r}))\;=\;x(x-1+\frac{1}{r})\;=\;x(\frac{rx-x+1}{r})=\frac{rx^2-x^2+1}{r}$
Поэтому для убодства деления поделим исходный многочлен на $-r\frac{rx^2-x^2+1}{r}\;=\;-rx^2+rx-x$, обозначим его за $D(x)$.
Домножили на $r$ чтобы не делить на дробь, домножили на $-1$, так как у исходного многочлена старший член со знаком минус, 
чтобы после деления ответ не был со знаком минус.
Теперь разделим многочлен на многочлен:
Первый член многочлена $P(x)$ это $-r^3x^4$ и делителя $D(x)$ $-rx^2$, делим: $\frac{-r^3x^4}{-rx^2}\;=\;r^2x^2$, обозначим его за $t_1(x)$.
Получим остаток $R_1(x)=P(x)-t_1(x)*D(x)\;=\;(-r^3x^4+2r^3x^3-(r^3+r^2)x^2+(r^2-1)x) - (-r^3x^4+r^3x^3-r^2x^3)\;=\;$
$\;=\;(r^3+r^2)x^3-(r^3+r^2)x^2+(r^2-1)x$
Первый член многочлена $R_1(x)$ это $(r^3+r^2)x^3$, делим на $D(x)$: $\frac{(r^3+r^2)x^3}{-rx^2}\;=\;-(r^2+r)x$, обозначим его за $t_2(x)$.
Получим остаток $R_2(x)=R_1(x)-t_2(x)*D(x)\;=\;((r^3+r^2)x^3-(r^3+r^2)x^2+(r^2-1)x)-(r^3x^3-r^3x^2+r^2x^3+rx^2)\;=\;-(r^2+r)x^2+(r^2-1)x$
Первый член многочлена $R_2(x)$ это $-(r^2+r)x^2$, делим на $D(x)$: $\frac{-(r^2+r)x^2}{-rx^2}\;=\;r+1$, обозначим его за $t_3(x)$.
Получим остаток $R_3(x)=R_2(x)-t_3(x)*D(x)\;=\;(-(r^2+r)x^2+(r^2-1)x)-(-(r^2-r)x^2+(r^2-1)x)=0$, тогда деление окончено, так как остаток равен нулю.
Получаем $P(x)\;=\;-rx(x-x^*)(t_1(x)+t_2(x)+t_3(x))\;=\;-rx(x-x^*)(r^2x^2-(r^2+r)x+r+1)$.
Теперь рассмотрим знак многочлена $t_1(x)+t_2(x)+t_3(x)$, обозначим его за $T(x)$: $t_1(x)+t_2(x)+t_3(x)\;=\;r^2x^2-(r^2+r)x+r+1$
Это парабола ветвями вверх, найдем ее минимальное значение, его можно вычислить по формуле $y_0\;=\;\frac{4ac-b^2}{4a}$
$y_0\;=\;\frac{4r^2(r+1)-(r^2+r)^2}{4r^2}\;=\;\frac{4r^2(r+1)-(r^4+2r^3+r^2)}{4r^2}\;=\;\frac{4(r+1)-(r^2+2r+1)}{4}\;=\;\frac{-r^2+2r+3}{4}\;=\;\frac{(3-r)(r+1)}{4}$
Теперь определим знак $y_0$, по условию $r\in(2;3)$, тогда $(3-r) > 0$ и $(r+1)>0$, тогда $y_0>0$ при любых данных $r$.
Так как $T(x)$ - парабола ветвями вверх, то $y_0$ - наименьшее значение, так как $y_0>0$ при данных $r$, то $T(x)>0$, значит знак $P(x)$  не зависит от $T(x)$
Значит знак $P(x)\;=\;-rx(x-x^*)*T(x)$ зависит только от $-rx(x-x^*)$, но $-rx < 0$ при $x\in(0;1)$, тогда знак зависит только от $x-x^*$.
Получаем, что $P(x)<0$ при $(x-x^*)>0$, $P(x)>0$ при $(x-x^*)<0$, $P(x)=0$ при $x=x^*$ или $x=0$.
Так как мы рассматривали $P(x)=f(f(x))-x$, то получаем $x>x^*\quad\Rightarrow\quad f(f(x))<x\quad x<x^*\quad\Rightarrow\quad f(f(x))>x$.

$x_{2n}>x^*\quad\Rightarrow\quad f(f(x_{2n}))<x_{2n} \quad\Rightarrow\quad x_{2n+2}<x_{2n} \quad\Rightarrow\quad x_{2(n+1)}<x_{2n}$, тогда $\{x_{2n}\}$ монотонно убывает.
$x_{2n+1}<x^*\quad\Rightarrow\quad f(f(x_{2n+1}))>x_{2n+1} \quad\Rightarrow\quad x_{2n+3}>x_{2n} \quad\Rightarrow\quad x_{2(n+1)+1}>x_{2n+1}$, тогда $\{x_{2n+1}\}$ монотонно возрастает.



```python
def f(x_, r_):
    return r_ * x_ * (1 - x_)


r = 2.9
x_0 = 0.79
n = 35

x = np.zeros(n + 1)
x[0] = x_0
for i in range(n):
    x[i + 1] = f(x[i], r)

even_indices = np.arange(0, n + 1, 2)
odd_indices = np.arange(1, n + 1, 2)

even_values = x[even_indices]
odd_values = x[odd_indices]

static_x = 1 - 1 / r
plt.figure(figsize=(9,4))

plt.axhline(static_x, color='gray', linestyle='--', label=f"x* = {static_x:.4f}")

plt.plot(even_indices, even_values, 'o-', label='Подпоследовательность $x_{2n}$')
plt.plot(odd_indices,  odd_values,  's-', label='Подпоследовательность $x_{2n+1}$')
plt.xlabel("n")
plt.ylabel("x_n")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
